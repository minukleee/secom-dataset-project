```python
top_40_features = top_40_corr.index
X = df[top_40_features]



print("--- Successfully selected the top 40 features ---")
print(f"Shape of the new feature set X: {X.shape}")



from sklearn.model_selection import train_test_split


y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0, stratify=y)

print("\n--- Train/Test split complete ---")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

smote = SMOTE(random_state=0)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

print(f"Original training set distribution: {Counter(y_train)}")
print(f"After SMOTE distribution: {Counter(y_train_smote)}")


smoteenn = SMOTEENN(random_state=0)
X_train_smoteenn, y_train_smoteenn = smoteenn.fit_resample(X_train, y_train)

print(f"Original training set distribution: {Counter(y_train)}")
print(f"After SMOTEENN distribution: {Counter(y_train_smoteenn)}")

```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix as sk_confusion_matrix


classifier = SVC(kernel='rbf')
classifier.fit(X_train_smote, y_train_smote)
y_pred = classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"SVC + SMOTE accuracy: {accuracy:.2f}")


y_pred_mapped1 = np.where(y_pred == 0, -1, y_pred)


cm = confusion_matrix(y_test, y_pred_mapped1)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[-1, 1], yticklabels=[-1, 1])
plt.title('SVC + SMOTE')
plt.xlabel('Predict')
plt.ylabel('True')
plt.show()
```
